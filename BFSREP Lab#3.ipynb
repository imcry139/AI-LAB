```python
# BFS_REG#.py
# Breadth-First Search implementation for Romanian cities (Arad → Bucharest)

from collections import deque

# Graph with distances (simplified Romania map)
graph = {
    'Arad': [('Sibiu', 140), ('Timisoara', 118), ('Zerind', 75)],
    'Sibiu': [('Arad', 140), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Pitesti', 97), ('Craiova', 146)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Oradea': [('Zerind', 71), ('Iasi', 151)],  # extra connections
    'Iasi': [('Oradea', 151), ('Neamt', 87), ('Vaslui', 92)],
    'Vaslui': [('Iasi', 92), ('Urziceni', 142)],
    'Urziceni': [('Vaslui', 142), ('Bucharest', 85), ('Hirsova', 98)],
    'Hirsova': [('Urziceni', 98), ('Eforie', 101)],
    'Eforie': [('Hirsova', 101)],
    'Neamt': [('Iasi', 87)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Urziceni', 85), ('Giurgiu', 90)],
    'Giurgiu': [('Bucharest', 90)]
}

def bfs_path(start, goal):
    visited = set()
    queue = deque([(start, [start], 0)])  # (current_node, path, cost)

    while queue:
        node, path, cost = queue.popleft()
        if node == goal:
            return path, cost
        if node not in visited:
            visited.add(node)
            for neighbor, distance in graph.get(node, []):
                if neighbor not in visited:
                    queue.append((neighbor, path + [neighbor], cost + distance))
    return None, None

# Driver code
start = 'Arad'
goal = 'Bucharest'
path, total_cost = bfs_path(start, goal)

print("BFS Path from", start, "to", goal, ":")
print(" → ".join(path))
print("Total Distance:", total_cost)
```
