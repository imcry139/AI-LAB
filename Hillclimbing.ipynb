```python
# Hill Climbing Algorithm

# Step 1: Define hill climbing function
def hill_climb():

    # Step 2: Start from initial state
    current = 0

    # Step 3: Repeat until no better state exists
    while True:

        # Step 4: Generate neighbor state
        next_state = current + 1

        # Step 5: Compare states using objective function
        # Objective function: maximize x^2

        if (next_state ** 2) > (current ** 2) and next_state <= 5:

            # Step 6: Move to better state
            current = next_state

        else:
            # Step 7: Stop if no improvement
            break

    # Step 8: Return best solution
    return current


# Step 9: Call algorithm
result = hill_climb()

# Step 10: Print result
print("Best State:", result)
```
