```python
# Parameterized Hill Climbing Algorithm

# Step 1: Define objective function
def objective_function(x):
    return x ** 2   # maximize x²


# Step 2: Define hill climbing function
def hill_climb(start, step_size, max_limit):

    # Initial state
    current = start

    # Repeat until no better neighbor exists
    while True:

        # Generate neighbor state
        next_state = current + step_size

        # Compare states using objective function
        if (
            objective_function(next_state)
            > objective_function(current)
            and next_state <= max_limit
        ):

            # Move to better state
            current = next_state

        else:
            # Stop if no improvement
            break

    # Return best solution
    return current


# Step 3: Parameters
start_state = 0
step = 1
limit = 5

# Step 4: Call algorithm
result = hill_climb(start_state, step, limit)

# Step 5: Print result
print("Best State:", result)
print("Best Value:", objective_function(result))
```
