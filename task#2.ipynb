**Task 2: Object-Oriented Design and Area Comparison**

1\. Classes for Trapezoid and Parallelogram

class Trapezoid:  
def \_\_init\_\_(self, a, b, h):  
self.a = a  
self.b = b  
self.h = h

> def area(self):  
> return 0.5 \* (self.a + self.b) \* self.h

class Parallelogram:  
def \_\_init\_\_(self, base, height):  
self.base = base  
self.height = height

> def area(self):
>
> return self.base \* self.height
